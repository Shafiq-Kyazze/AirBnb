```python
def status(dt):
    stat = pd.DataFrame(index=dt.columns)
    stat['percentage_of_missing_values'] = (dt.isna().sum()/len(dt)) * 100
    stat['sample'] = dt.iloc[10]
    stat['data type'] = dt.dtypes
    return stat.sort_values('percentage_of_missing_values',ascending=False).head(-1)
status(df)
```

### Cleaning and preprocessing individual columns

##### House rules and host about 
These columns have alot of information missing values so they will be dropped

```python
df.drop(['house_rules','host_about'],axis=1,inplace=True)
```

##### Security deposit
This column has over a third of its values missing. 
The missing values will be assumed to requireno deposit

```python
df.security_deposit.unique()
```

```python
df.security_deposit.fillna(0,inplace=True)
df.security_deposit.isna().sum()
```

```python
df.security_deposit = df.security_deposit.astype('str') #do this so that we can strip each cell of $
df.security_deposit = df.security_deposit.str.replace('$','')
df.security_deposit = df.security_deposit.str.replace(',','') #get rid of the comma after 1,000.00
df.security_deposit = df.security_deposit.astype('float64')
```

##### Host response time
The missing response times will be labelled as unknown


```python
df.host_response_time.unique()
```

```python
df.host_response_time.fillna('unknown',inplace=True)
```

#### host response rate
Over a third of rows have missing host response rates. This column will be kept since host response rates contribute towards customer reviews.
The ratings will be grouped into groups of 10s

```python
df.host_response_rate.unique()
```

```python
df.host_response_rate.fillna('-1',inplace=True)
```

```python
df.host_response_rate = df.host_response_rate.str.replace('%','')
df.host_response_rate = df.host_response_rate.astype('int64')
df.host_response_rate = df.host_response_rate.apply(lambda x: x//10 * 10)
df.host_response_rate = df.host_response_rate.replace(-10,'unknown')
```

#### Review scores rating
Almost a third of the rows have missing values in this column. The missing values will be labelled as unknown
* To keep the datatype as an integer, unknown ratings are represented as -1

```python
df.review_scores_rating.unique()
```

```python
df.review_scores_rating.fillna(-1,inplace=True)  
#df.review_scores_rating = df.review_scores_rating.replace(-1,'unknown')
```

#### review_scores_accuracy
Almost a third of the rows have missing values in this column. The missing values will be labelled as unknown
* To keep the datatype as an integer, unknown ratings are represented as -1

```python
df.review_scores_accuracy.value_counts()
```

```python
df.review_scores_accuracy.fillna(-1,inplace=True)  
#df.review_scores_rating = df.review_scores_rating.replace(-1,'unknown')
```

#### Description

####  bedrooms,beds,bathrooms
These columns' missing values will be filled witht he mean of the columns

```python
for col in ['beds','bedrooms','bathrooms']:
    df[col].fillna(df[col].mean(),inplace=True)
    df[col] = df[col].astype('float64')
```

#### Cleaning fee
Missing values will be filled with the median cleaning fee

```python
df.cleaning_fee.unique()
```

```python
df.cleaning_fee = df.cleaning_fee.str.replace('$','') 
df.cleaning_fee =df.cleaning_fee.str.replace(',','') 
df.cleaning_fee.fillna(df.cleaning_fee.median(),inplace=True)
df.cleaning_fee = df.cleaning_fee.astype('float64')
```

#### Price

```python
df.price = df.price.str.strip('$')
df.price = df.price.str.replace(',','')
df.price=df.price.astype(float)
```

#### Amenities
Amenities will be reduced to just the 10 most important amenities which are based on 2 articles from the Airbnb website and  about the top 10 amenities.[Link](https://www.airbnb.co.uk/resources/hosting-homes/a/the-best-amenities-to-offer-right-now-203) ,[Link2](https://moneywise.com/a/top-things-guests-want-in-an-airbnb).
These will include: Wi-Fi,air-conditioning,Kitchen,parking, friendly pet policy,Jacuzzi,washing machine,heating,TV,swimming pool

```python
df.amenities[:1].values
```

```python
#Creating a list of all the amenities provided by hosts
Amenities_list = list(df.amenities)
Amenities_list_as_str = ' '.join(Amenities_list)
Amenities_list_as_str = Amenities_list_as_str.replace('{','')
Amenities_list_as_str = Amenities_list_as_str.replace('}','')
Amenities_list_as_str = Amenities_list_as_str.replace('"' ,'')
Amenities_listt = list(Amenities_list_as_str.split(','))
Amenities_listt = list(set(Amenities_listt))
sorted(Amenities_listt)
```

10 Columns of amenities are going to be created based on the articles metioned above.
These include:Wi-Fi,air-conditioning,Kitchen,parking, friendly pet policy,Jacuzzi/spa,washing machine,heating,TV,swimming pool

```python
df.loc[df.amenities.str.lower().str.contains('wifi|wi-fi'), 'WiFi'] = 1
df.loc[df.amenities.str.lower().str.contains('air conditioning'), 'Air_conditioning'] = 1
df.loc[df.amenities.str.lower().str.contains('kitchen'), 'Kitchen'] = 1
df.loc[df.amenities.str.lower().str.contains('parking'), 'Parking'] = 1
df.loc[df.amenities.str.lower().str.contains('pets allowed|pets'), 'Pet_friendliness'] = 1
df.loc[df.amenities.str.lower().str.contains('jacuzzi|spa'), 'Jacuzzi'] = 1
df.loc[df.amenities.str.lower().str.contains('washing machine|washer'), 'Washer'] = 1
df.loc[df.amenities.str.lower().str.contains('heating'), 'Heating'] = 1
df.loc[df.amenities.str.lower().str.contains('tv'), 'TV'] = 1
df.loc[df.amenities.str.lower().str.contains('swimming pool|pool'), 'Swimming pool'] = 1
```

```python
#filling missing values of 10 amenity columns with 0 and dropping the amenities column
df.fillna(0,inplace=True)
df.drop('amenities',axis=1,inplace=True)
```

#### Host since
Going to be changed from object to date-time

```python
df['host_since'] = pd.to_datetime(df.host_since)
```

#### Duration of being a host
A new column that shows the total duration a user has been a host
The date used as the final date is the date the data was compiled. This date is 09/01/2020 which was obtained from [inside Airbnb website](http://insideairbnb.com/get-the-data.html)

```python
Final_date = pd.to_datetime('09/01/2020')
df['Duration_as_host'] = Final_date - df.host_since
df.drop( 'host_since',axis=1,inplace=True)
df.Duration_as_host
```

```python
#This column was then converted from days to years
import numpy as np
df.Duration_as_host = df.Duration_as_host/np.timedelta64(1,'Y')
df.rename(columns={'Duration_as_host':'Years_as_host'},inplace=True)
```
